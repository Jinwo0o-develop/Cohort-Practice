```python
%load_ext autoreload

%autoreload 2

import config
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

config.set_korean_font()
```

```python
file_name = pd.read_csv("tbsh_gyeonggi_day_202507_이천시.csv", encoding='utf-8')
df = file_name

# config에서 매핑자료로 매핑
df['age'] = df['age'].map(config.AGE_MAP)
df['hour'] = df['hour'].map(config.TIME_SHORT_MAP)
df['day'] = df['day'].map(config.DAY_MAP)

# 카테고리화
df['sex'] = df['sex'].astype('category')
df['card_tpbuz_cd'] = df['card_tpbuz_cd'].astype('category')
df['card_tpbuz_nm_1'] = df['card_tpbuz_nm_1'].astype('category')
df['card_tpbuz_nm_2'] = df['card_tpbuz_nm_2'].astype('category')

df.info()
```

```python
# **MISSION 2: 세대별 소비 트렌드 비교 (Cohort Analysis)**
## 청년층(2030) / 중년층(4050) / 노년층60+ 분리 -> 세그멘테이션
### 2030 / 20대 / 30대 분리 시각화
#### 필요할 때 꺼내 쓰기

import my_plotting as mp


age_2030 = df[df['age'].isin(['20대','30대'])]
age_4050 = df[df['age'].isin(['40대','50대'])]
age_60 = df[df['age'].isin(['60대','70대','80대','90대','100세 이상'])]

age_20 = df[df['age'].isin(['20대'])]
age_30 = df[df['age'].isin(['30대'])]
age_40 = df[df['age'].isin(['40대'])]
age_50 = df[df['age'].isin(['50대'])]


data_2030 = age_2030.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_4050 = age_4050.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_60 = age_60.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_20 = age_20.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_30 = age_30.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_40 = age_40.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)
data_50 = age_50.groupby(['card_tpbuz_nm_1'])['amt'].sum().sort_values(ascending=False)


fig, axes = plt.subplots(4, 2, figsize=(12,25))

# axes로 차원분할보다 ax를 정한 후 한가화 하는게 더 나을 수도
# 20대에는 소매/유통의 반 이상이 음식임 (2030 / 4050 단위로 봤을 떄에는 절반을 넘지 못 함. )
my_plotting.draw_top_categories(data_2030, '2030 업종별 매출총액', axes[0][0])
my_plotting.draw_top_categories(data_20, '20 업종별 매출총액', axes[0][1])
my_plotting.draw_top_categories(data_30, '30 업종별 매출총액', axes[1][0])
my_plotting.draw_top_categories(data_4050, '4050 업종별 매출총액', axes[1][1])
my_plotting.draw_top_categories(data_40, '40 업종별 매출총액', axes[2][0])
my_plotting.draw_top_categories(data_50, '50 업종별 매출총액', axes[2][1])
my_plotting.draw_top_categories(data_60, '60세 이상 업종별 매출총액', axes[3][0])
```
